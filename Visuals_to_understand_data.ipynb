```python
import matplotlib.pyplot as plt
import numpy as np

# Example data: Student scores
student_in_a_class = [50, 152, 160, 165, 170, 175, 180, 185, 195, 100]  # Includes an outlier (110)

# Create Box Plot
plt.figure(figsize=(8, 5))
box = plt.boxplot(scores, vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))

# Calculate statistics manually
quartiles = np.percentile(student_in_a_class, [25, 50, 75])
q1, median, q3 = quartiles
iqr = q3 - q1
lower_whisker = max(min(scores), q1 - 1.5 * iqr)
upper_whisker = min(max(scores), q3 + 1.5 * iqr)

# Annotate components
plt.text(lower_whisker, 1.05, f"Lower Whisker ({lower_whisker:.1f})", fontsize=10, ha="center", color="green")
plt.text(upper_whisker, 1.05, f"Upper Whisker ({upper_whisker:.1f})", fontsize=10, ha="center", color="green")
plt.text(median, 1.2, f"Median ({median})", fontsize=10, ha="center", color="red")
plt.text(q1, 1.2, f"Q1 ({q1})", fontsize=10, ha="center", color="blue")
plt.text(q3, 1.2, f"Q3 ({q3})", fontsize=10, ha="center", color="blue")

# Highlight outlier(s)
outliers = [score for score in scores if score > upper_whisker or score < lower_whisker]
for outlier in outliers:
    plt.text(outlier, 0.95, f"Outlier ({outlier})", fontsize=10, ha="center", color="purple")

# Final formatting
plt.title("Annotated Box Plot of Exam Scores")
plt.xlabel("Scores")
plt.yticks([])  # Remove y-axis ticks for clarity
plt.show()
```

**SKEWNESS**

```python
from scipy.stats import skew

# Example data
data = [1, 2, 2, 3, 4, 10]  # Positively skewed data

# Calculate skewness
skewness_value = skew(data)
print(f"Skewness: {skewness_value}")

# Plot histogram to visualize skewness
plt.hist(data, bins=5, color='lightgreen', edgecolor='black')
plt.title("Histogram Showing Skewness")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import skew

# Generate skewed data
np.random.seed(42)
positive_skew_data = np.random.exponential(scale=2, size=1000)  # Positively skewed
negative_skew_data = -1 * np.random.exponential(scale=2, size=1000) + 10  # Negatively skewed

# Calculate skewness
positive_skewness = skew(positive_skew_data)
negative_skewness = skew(negative_skew_data)

# Create subplots
plt.figure(figsize=(12, 6))

# Positive Skewness
plt.subplot(1, 2, 1)
plt.hist(positive_skew_data, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(positive_skew_data), color='red', linestyle='dashed', linewidth=1.5)
plt.text(np.mean(positive_skew_data), 80, f"Mean = {np.mean(positive_skew_data):.2f}", color='red', ha='center')
plt.title(f"Positively Skewed Data\nSkewness = {positive_skewness:.2f}")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Negative Skewness
plt.subplot(1, 2, 2)
plt.hist(negative_skew_data, bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(negative_skew_data), color='red', linestyle='dashed', linewidth=1.5)
plt.text(np.mean(negative_skew_data), 80, f"Mean = {np.mean(negative_skew_data):.2f}", color='red', ha='center')
plt.title(f"Negatively Skewed Data\nSkewness = {negative_skewness:.2f}")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

**KURTOSIS**

```python
from scipy.stats import kurtosis

# Example data
data = [2, 3, 3, 4, 5, 5, 5, 5, 5, 6, 6, 7]  # High kurtosis data

# Calculate kurtosis
kurtosis_value = kurtosis(data)
print(f"Kurtosis: {kurtosis_value}")

# Plot histogram to visualize kurtosis
plt.hist(data, bins=5, color='coral', edgecolor='black')
plt.title("Histogram Showing Kurtosis")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Generate data for High and Low Kurtosis
np.random.seed(42)

# High Kurtosis: Many data points close to the mean, heavy tails
high_kurtosis_data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution
high_kurtosis_data = np.append(high_kurtosis_data, [10, -10, 15, -15])  # Adding outliers for heavy tails

# Low Kurtosis: Data spread more evenly with lighter tails
low_kurtosis_data = np.random.uniform(low=-3, high=3, size=1000)  # Uniform distribution

# Calculate kurtosis
high_kurt = kurtosis(high_kurtosis_data)
low_kurt = kurtosis(low_kurtosis_data)

# Plot histograms
plt.figure(figsize=(12, 6))

# High Kurtosis Plot
plt.subplot(1, 2, 1)
plt.hist(high_kurtosis_data, bins=30, color='lightblue', edgecolor='black', alpha=0.7)
plt.title(f"High Kurtosis (Leptokurtic)\nKurtosis = {high_kurt:.2f}")
plt.axvline(np.mean(high_kurtosis_data), color='red', linestyle='dashed', linewidth=1.5)
plt.text(np.mean(high_kurtosis_data), 50, f"Mean = {np.mean(high_kurtosis_data):.2f}", color='red', ha='center')
plt.xlabel("Value")
plt.ylabel("Frequency")

# Low Kurtosis Plot
plt.subplot(1, 2, 2)
plt.hist(low_kurtosis_data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title(f"Low Kurtosis (Platykurtic)\nKurtosis = {low_kurt:.2f}")
plt.axvline(np.mean(low_kurtosis_data), color='red', linestyle='dashed', linewidth=1.5)
plt.text(np.mean(low_kurtosis_data), 50, f"Mean = {np.mean(low_kurtosis_data):.2f}", color='red', ha='center')
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
